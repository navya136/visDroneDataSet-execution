# Real Time Detections

```python
cap = cv2.VideoCapture(0)  # this code to access the web camera
while cap.isOpened():  
    ret, frame = cap.read() # it captures the images from the camera
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render())) #this code will name the popup screen with 'yolo'
    
    if cv2.waitKey(10) & 0xFF == ord('q'): # this is for exit the camera
        break
cap.release()# it will release webcam
cv2.destroyAllWindows()
```
